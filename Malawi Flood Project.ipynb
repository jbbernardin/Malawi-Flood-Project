# Modeling

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from xgboost import XGBRegressor

from scipy.stats import randint as sp_randint
from scipy.stats import uniform 

from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
```

```python
X = precip_2015
y = target

#Finding the best parameters for the ridge model
ridge = Ridge()
search_parameters_space_random = {"alpha": uniform(loc=0.1, scale=10), "fit_intercept": [True, False]}
random = RandomizedSearchCV(estimator=ridge, param_distributions=search_parameters_space_random,scoring="r2", n_jobs=-1, n_iter=50)
```

```python
#evaluation function
kfold  = KFold(n_splits=10, shuffle = True, random_state=42)

def evaluate(estimator):
    cv_results = cross_validate(estimator,
                    X=X,
                    y=y,
                    scoring="neg_mean_absolute_error",
                          n_jobs=-1, cv=kfold,
                     return_train_score=True)
    return pd.DataFrame(cv_results).abs().mean().to_dict()
```

```python
#Evaluate the models
scores = {}

scores['linear'] = evaluate(LinearRegression())
scores['random'] = evaluate(random)
scores['lasso'] = evaluate(Lasso())
scores['elastic'] = evaluate(ElasticNet())
scores['tree'] = evaluate(DecisionTreeRegressor())
scores['forrest'] = evaluate(RandomForestRegressor())
scores['gradient'] = evaluate(GradientBoostingRegressor())
scores['ada'] = evaluate(AdaBoostRegressor())
scores['xg'] = evaluate(XGBRegressor())

pd.DataFrame.from_dict(scores).sort_values('test_score', axis=1).T
```

### So the best three models are forrest, tree, and xgboost

```python
#stack the three best models, use bagging, then estimate
from sklearn.ensemble import StackingRegressor, BaggingRegressor

stacking = StackingRegressor(
    estimators=[
        ('forrest',RandomForestRegressor()),
        ('tree',DecisionTreeRegressor()),
        ('xg',XGBRegressor())], 
    final_estimator=XGBRegressor())

bagging10 = BaggingRegressor(n_estimators=10)
bagging100 = BaggingRegressor(n_estimators=100)

scores['stacking_top3'] = evaluate(stacking)
scores['bagging_10'] = evaluate(bagging10)
scores['bagging_100'] = evaluate(bagging100)

pd.DataFrame.from_dict(scores).sort_values('test_score', axis=1).T
```

## It seems as though Random Forrest Regression and Bagging with 100 estimators are nearly identical, both in test score and train score, so we proceed with Random Forrest Regression since it posesses a marginally better test score
